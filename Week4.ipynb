```python
#Week 4
def add_numbers(a, b):
    return a + b

result = add_numbers(5, 3)
print(result)
```

```python
def greet(name, age):
    print(f"Hello {name}, you are {age} years old")

greet("Ankita", 25)
```

```python
greet(age=25, name="Ankita")
```

```python
def greet(name, country="USA"):
    print(f"{name} is from {country}")

greet("Ankita")
greet("Aman", "India")
```

```python
def add_all(*numbers):
    print(sum(numbers))

add_all(1, 2, 3, 4)
```

```python
def print_info(**details):
    print(details)

print_info(name="Ankita", age=25, city="Sydney")
```

```python
square = lambda x: x * x
print(square(5))
```

```python
items = [(1, 'b'), (3, 'a'), (2, 'c')]
items.sort(key=lambda x: x[1])
print(items)
```

```python
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x*x, nums))
print(squares)
```

```python
even = list(filter(lambda x: x%2==0, nums))
print(even)
```

```python
from functools import reduce
total = reduce(lambda x, y: x + y, nums)
print(total)
```

```python
try:
    a = int(input("Enter number: "))
    print(10 / a)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid conversion")
finally:
    print("Program ended")
```

```python
# Create a new file and write text
with open("data.txt", "w") as file:
    file.write("This is Week 4 file handling.\n")
    file.write("Learning read, write, append operations.\n")

print("File created successfully!")
```

```python
with open("data.txt", "r") as file:
    content = file.read()
    print("Reading full file:")
    print(content)
```

```python
with open("data.txt", "r") as file:
    for line in file:
        print("Line:", line.strip())
```

```python
with open("data.txt", "a") as file:
    file.write("This text is added using append mode.\n")

print("Content appended!")
```

```python
with open("data.txt", "r") as file:
    print(file.read())
```

```python
items = ["Apple", "Banana", "Cherry"]

# Write list items
with open("fruits.txt", "w") as file:
    for item in items:
        file.write(item + "\n")

print("List saved to fruits.txt!")
with open("fruits.txt", "r") as file:
    print(file.read())
```

```python
import os

if os.path.exists("data.txt"):
    print("data.txt exists!")
else:
    print("File not found!")
```

```python
import os

if os.path.exists("fruits.txt"):
    os.remove("fruits.txt")
    print("fruits.txt deleted.")
else:
    print("File does not exist!")
```
