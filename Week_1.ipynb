```python
# Day 2: Loops and List Operations

# Reset list
fruits = ["apple", "banana", "cherry", "mango"]

# List operations
fruits.append("orange")        # Add
fruits.remove("banana")        # Remove

print("\nUpdated fruits list:")
for fruit in fruits:
    print(fruit)
```

```python
# Multiplication Table 1-10
print("\nMultiplication Table (1-10):")
for i in range(1, 6):
    for j in range(1, 11):
        print(f"{i} x {j} = {i*j}")
    print("------")
```

```python
# Day 3: Functions

# Factorial function
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

print(f"\nFactorial of 5 is {factorial(5)}")
```

```python
# Sum of numbers function
def sum_numbers(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total

print(f"Sum of numbers 1 to 10 is {sum_numbers(10)}")
```

```python
# Day 4: File Handling

# Write contacts to a file
with open("contacts.txt", "w") as file:
    file.write("Ankita\nJohn\nSara\n")

# Read and print contacts
with open("contacts.txt", "r") as file:
    data = file.read()
    print("\nContacts from file:")
    print(data)
```

```python
# Day 5: Error Handling

try:
    x = int(input("Enter a number: "))
    y = int(input("Enter another number: "))
    print(f"Division result: {x / y}")
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input!")
finally:
    print("Execution completed.")
```

```python
# Day 6: Simple To-Do List

tasks = []

def add_task(task):
    tasks.append(task)
    print(f"Added task: {task}")

def view_tasks():
    print("\nYour tasks:")
    for t in tasks:
        print("-", t)

add_task("Buy groceries")
add_task("Do Python practice")
view_tasks()
```

```python
# Day 7: Contact Manager

contacts = []

def add_contact(name):
    contacts.append(name)

def view_contacts():
    print("\nContacts List:")
    for contact in contacts:
        print("-", contact)

# Adding sample contacts
add_contact("Ankita")
add_contact("John")
add_contact("Sara")
view_contacts()
```
