```python
# Week 2 – Python Functions & Code Structure
# DAY 1 – Functions Basics

# 1. Simple Function
def greet():
    print("Hello! Welcome to Week 2.")
greet()


# 2. Function with Parameters
def welcome(name):
    print(f"Hello {name}, great to see you learning Python!")
welcome("Ankita")


# 3. Function with Return Value
def add(a, b):
    return a + b
print("Addition:", add(15, 5))


# 4. Default Arguments
def profile(name, country="Australia"):
    print(f"Name: {name}, Country: {country}")

profile("Ankita")
profile("John", "India")


# 5. Function Scope Example
x = 10  # global

def show_scope():
    x = 20  # local
    print("Inside function:", x)

show_scope()
print("Outside function:", x)
```

```python
# 1. Multiply 3 numbers
def multiply(a, b, c):
    return a * b * c

# 2. Even or odd
def check_even(num):
    return "Even" if num % 2 == 0 else "Odd"

# 3. Largest of 3 numbers
def largest(a, b, c):
    return max(a, b, c)

# 4. Reverse a string
def reverse_string(s):
    return s[::-1]

# 5. Count vowels
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

print(multiply(2, 3, 4))
print(check_even(7))
print(largest(10, 4, 7))
print(reverse_string("ankita"))
print(count_vowels("Hello World"))
```

```python
# DAY 2 – Intermediate Functions

# 1. variable number of arguments
def total_sum(*numbers):
    return sum(numbers)

print("Total: ", total_sum(1,2,3,4,5))


# 2. keyword arguments
def user_details(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

user_details(name="Ankita", age=22, country="Australia")


# 3. Lambda Functions
square = lambda x: x*x
print("Square of 5:", square(5))


# 4. Nested Functions
def outer_function(msg):
    def inner():
        print("Inner message:", msg)
    inner()

outer_function("Learning Python!")


# 5. Passing a function as argument
def execute(fun, value):
    print("Executing function:", fun(value))

execute(square, 8)
```

```python
# 1. Write a lambda to find cube
cube = lambda x: x**3
print(cube(3))

# 2. Create a function that applies any function to a number
def apply(func, num):
    return func(num)

print(apply(cube, 4))

# 3. Create a function with *args that returns average
def average(*nums):
    return sum(nums) / len(nums)

print(average(3,4,5,6))
```

```python
# Week 2 Mini Project – Student Grade Calculator

def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 75:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "D"

def student_report():
    name = input("Enter student name: ")
    marks = int(input("Enter marks (0–100): "))

    grade = calculate_grade(marks)

    print("\n--- Student Report ---")
    print("Name:", name)
    print("Marks:", marks)
    print("Grade:", grade)

student_report()
```
